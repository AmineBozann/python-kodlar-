```python
# ğŸŒ¦ï¸ HAVA DURUMU TAHMÄ°N ARACI (SimÃ¼lasyon)
# Bu program bir ÅŸehir adÄ± alÄ±r, rastgele hava verileri Ã¼retir
# ve hava durumuna dair tahmini analiz eder.

import random               # Rastgele sÄ±caklÄ±k ve nem verileri oluÅŸturmak iÃ§in
import datetime             # GÃ¼n bilgisi iÃ§in
import matplotlib.pyplot as plt  # Grafik Ã§izmek iÃ§in

# ğŸ”¹ 1. AdÄ±m: Rastgele hava durumu verisi Ã¼reten fonksiyon
def hava_verisi_uret():
    """
    Rastgele sÄ±caklÄ±k, nem ve rÃ¼zgar hÄ±zÄ± deÄŸerleri Ã¼retir.
    GerÃ§ek API yerine simÃ¼lasyon amaÃ§lÄ± kullanÄ±lÄ±r.
    """
    sicaklik = random.uniform(-5, 40)   # -5Â°C ile 40Â°C arasÄ±
    nem = random.randint(10, 100)       # %10 ile %100 arasÄ±
    ruzgar = random.uniform(0.5, 15.0)  # 0.5â€“15 m/s arasÄ±
    return sicaklik, nem, ruzgar

# ğŸ”¹ 2. AdÄ±m: Tahmin algoritmasÄ±
def hava_tahmin_et(sicaklik, nem, ruzgar):
    """
    Hava verilerine gÃ¶re basit bir kural tabanlÄ± tahmin Ã¼retir.
    """
    if nem > 80 and sicaklik < 20:
        return "ğŸŒ§ï¸ YaÄŸmurlu"
    elif sicaklik > 30 and nem < 50:
        return "â˜€ï¸ GÃ¼neÅŸli"
    elif ruzgar > 10:
        return "ğŸ’¨ RÃ¼zgarlÄ±"
    elif sicaklik < 0:
        return "â„ï¸ KarlÄ±"
    else:
        return "â›… ParÃ§alÄ± Bulutlu"

# ğŸ”¹ 3. AdÄ±m: GÃ¼nlÃ¼k 7 tahminlik veri simÃ¼lasyonu
def haftalik_tahmin_uret():
    """
    7 gÃ¼nlÃ¼k rastgele hava verisi Ã¼retir ve tahminleri dÃ¶ndÃ¼rÃ¼r.
    """
    gunler = []
    sicakliklar = []
    tahminler = []

    for i in range(7):
        gun = (datetime.date.today() + datetime.timedelta(days=i)).strftime("%a")
        sicaklik, nem, ruzgar = hava_verisi_uret()
        tahmin = hava_tahmin_et(sicaklik, nem, ruzgar)

        gunler.append(gun)
        sicakliklar.append(sicaklik)
        tahminler.append((gun, sicaklik, nem, ruzgar, tahmin))

    return gunler, sicakliklar, tahminler

# ğŸ”¹ 4. AdÄ±m: SonuÃ§larÄ± ekrana yazdÄ±r
def tahmin_yazdir(sehir, tahminler):
    """
    GÃ¼nlÃ¼k tahminleri tablo gibi ekrana yazdÄ±rÄ±r.
    """
    print(f"\nğŸ“ {sehir.upper()} iÃ§in 7 GÃ¼nlÃ¼k Hava Durumu Tahmini:\n")
    print(f"{'GÃ¼n':<10} {'SÄ±caklÄ±k(Â°C)':<15} {'Nem(%)':<10} {'RÃ¼zgar(m/s)':<15} {'Durum'}")
    print("-" * 60)
    for gun, sicaklik, nem, ruzgar, tahmin in tahminler:
        print(f"{gun:<10} {sicaklik:>7.1f}{'Â°C':<8} {nem:<10} {ruzgar:<12.1f} {tahmin}")

# ğŸ”¹ 5. AdÄ±m: Grafik Ã§izen fonksiyon
def grafik_ciz(gunler, sicakliklar, sehir):
    """
    7 gÃ¼nlÃ¼k sÄ±caklÄ±k deÄŸiÅŸimini grafikle gÃ¶sterir.
    """
    plt.figure(figsize=(8,4))
    plt.plot(gunler, sicakliklar, marker="o", linewidth=2)
    plt.title(f"{sehir} - 7 GÃ¼nlÃ¼k SÄ±caklÄ±k Tahmini")
    plt.xlabel("GÃ¼n")
    plt.ylabel("SÄ±caklÄ±k (Â°C)")
    plt.grid(True)
    plt.show()

# ğŸ”¹ 6. AdÄ±m: Ana program
def ana_program():
    """
    KullanÄ±cÄ±dan ÅŸehir ismini alÄ±r, tahminleri Ã¼retir, sonucu gÃ¶sterir.
    """
    print("ğŸŒ¦ï¸ HAVA DURUMU TAHMÄ°N ARACI ğŸŒ¦ï¸")
    sehir = input("Tahmin almak istediÄŸiniz ÅŸehri girin: ")

    gunler, sicakliklar, tahminler = haftalik_tahmin_uret()
    tahmin_yazdir(sehir, tahminler)
    grafik_ciz(gunler, sicakliklar, sehir)

# ğŸ”¹ 7. AdÄ±m: ProgramÄ± Ã§alÄ±ÅŸtÄ±r
if __name__ == "__main__":
    ana_program()
```

```python
# ğŸŒ¦ï¸ HAVA DURUMU TAHMÄ°N UYGULAMASI
# KullanÄ±cÄ±dan ÅŸehir alÄ±r, OpenWeatherMap API'den hava durumu bilgisini Ã§eker
# Hava durumuna gÃ¶re kÄ±yafet Ã¶nerisi yapar

import requests  # HTTP isteÄŸi yapmak iÃ§in
import json      # JSON verisini iÅŸlemek iÃ§in

# ğŸŒ API bilgilerini tanÄ±mlÄ±yoruz
API_KEY = "demo"  # Kendi API anahtarÄ±nÄ± buraya eklemelisin (https://openweathermap.org/api)
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

def hava_durumu_getir(sehir):
    """
    Bu fonksiyon, verilen ÅŸehir iÃ§in hava durumu verisini getirir.
    OpenWeatherMap API'sini kullanÄ±r.
    """
    params = {
        "q": sehir,
        "appid": API_KEY,
        "lang": "tr",
        "units": "metric"  # sÄ±caklÄ±k Celsius olsun
    }
    response = requests.get(BASE_URL, params=params)

    if response.status_code == 200:
        return response.json()
    else:
        print("âŒ Åehir bulunamadÄ± veya API hatasÄ±.")
        return None

def kiyafet_onerisi(sicaklik, durum):
    """
    Hava sÄ±caklÄ±ÄŸÄ± ve duruma gÃ¶re Ã¶neri verir.
    """
    if "yaÄŸmur" in durum.lower():
        return "Åemsiye almayÄ± unutma â˜”"
    elif sicaklik < 5:
        return "KalÄ±n giyin, montunu giy ğŸ§¥"
    elif 5 <= sicaklik < 15:
        return "Hafif bir ceket yeterli olur ğŸ§¢"
    elif 15 <= sicaklik < 25:
        return "Hava gÃ¼zel, tiÅŸÃ¶rt giyebilirsin ğŸ‘•"
    else:
        return "OldukÃ§a sÄ±cak! Bol su iÃ§ ğŸ¥µ"

def ana_program():
    """
    Ana fonksiyon: kullanÄ±cÄ±dan ÅŸehir alÄ±r, hava durumu ve Ã¶neri verir.
    """
    print("ğŸŒ¦ï¸ HAVA DURUMU TAHMÄ°N UYGULAMASINA HOÅ GELDÄ°N ğŸŒ¦ï¸")
    sehir = input("Bir ÅŸehir adÄ± gir: ")

    veri = hava_durumu_getir(sehir)

    if veri:
        sicaklik = veri["main"]["temp"]
        durum = veri["weather"][0]["description"]
        print(f"\nğŸ“ {sehir.title()} iÃ§in hava durumu:")
        print(f"ğŸŒ¡ï¸ SÄ±caklÄ±k: {sicaklik}Â°C")
        print(f"ğŸŒ¤ï¸ Durum: {durum}")
        print(f"ğŸ‘• Ã–neri: {kiyafet_onerisi(sicaklik, durum)}")
    else:
        print("Veri alÄ±namadÄ±, lÃ¼tfen tekrar deneyin.")

# ProgramÄ± Ã§alÄ±ÅŸtÄ±r
if __name__ == "__main__":
    ana_program()
```
